```python
# Install necessary libraries
!pip install netCDF4 matplotlib
# Import libraries
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import requests
# Define the URL of the NetCDF file
url = "https://drive.google.com/uc?export=download&id=12X2Ui6nDRdPVPHsrT2uBvO_SKKIantnA"
# Download the file
response = requests.get(url)
with open('Suelo_Cobertura.nc', 'wb') as file:
    file.write(response.content)
# Open the NetCDF file
dataset = nc.Dataset('Suelo_Cobertura.nc')
# List the variables available in the dataset
print(dataset.variables.keys())
# Assuming that the first variable in the dataset is what you want to plot
variable_name = list(dataset.variables.keys())[0]
data = dataset.variables[variable_name][:]
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
plt.imshow(data[ :, :], cmap='viridis')  # Adjust indexing as per the dimensions of your data
plt.gca().invert_yaxis()  # Invert the y-axis
plt.colorbar()
plt.title('Suelo_Cobertura')
plt.show()
```
