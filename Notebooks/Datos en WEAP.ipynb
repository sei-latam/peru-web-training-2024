Este ejercicio consiste en ajustar el modelo para analizar la distribucion del agua bajo una condicion de disponibilidad hidrica especifica, mostrada en la figura previa.

Una vez concluida con el cargado de datos, se puede pasar la opción de "Resultados" en el menu izquierdo, con el cual WEAP realizar una simulacion con los datos, bajo la suencia siguiente. WEAP en **"Resultados"** puede mostrar varios reportes en base a la cantidad de datos suministrados, en nuestro caso no enfocaremos en dos reportes de **Demanda**, la **"Unsatisfied Demanda"**y **"Coverage"**, como se muestra a continuacion:

<img src="https://drive.google.com/uc?export=view&id=1A4bN0P2uyU3Qs4VPzlVN87ugfy7Xiu_J" alt="Drawing"  width="1000" style="width: 100px;"/>

Como se puede observar, WEAP realizo la distribución de agua generando una cobertura de 100% para la demanda poblacional, y de 83% de la demanda agricola, esto se debe a los numeros en parentesis mostrados al momento de crear los nodos de demanda. En esa parte para la agricultura se definio un valor "2" para este sitio, y para poblacional se dejo el valor por defecto "1". Los valores de cobertura y por lo tanto del deficit de suministro "Unsatisfied Demand" estimados son consecuencia de este ajuste del modelo.
