## **Algunos modelos hidrológicos**

## *Dynamic Water Balance - DWB*
Un estudio empírico realizado por [Budyko (1961)](https://doi.org/10.1080/00385417.1961.10770) estima el balance hídrico a largo plazo introduciendo una relación simple de oferta y demanda entre la energía y el agua superficial disponible a largo plazo. Este concepto, también conocido como marco de Budyko, se ha utilizado para construir modelos de balance hídrico. La importancia de estos modelos de balance hídrico radica en el hecho de que generalmente explican los aspectos fundamentales de un proceso hidrológico de manera suficiente con un número relativamente pequeño de entradas y parámetros del model.

[*Curva Budyko: Oferta-Demanda*](https://www.gloh2o.org/mswep/)
<img src="https://drive.google.com/uc?export=view&id=1CEsVdlq8_bqVb1g7p5_3M4MnsdD-Tz-q" alt="Drawing"  width="800" style="width: 100px;"/>

Para ampliar estos modelos de balance hídrico basados ​​en Budyko a escalas temporales más precisas es necesario incorporar procesos adicionales. [Zhang et al. (2008)](https://doi.org/10.1016/j.jhydrol.2008.07.021) argumentaron que la variación de las precipitaciones, la evapotranspiración potencial y el almacenamiento de agua deben tenerse en cuenta al modelar el balance hídrico en escalas temporales subanuales. En respuesta, propusieron un modelo de balance hídrico subanual con una estructura de múltiples capas para incluir el control del almacenamiento e imitar el proceso de partición física.

[*Esquematica del modelo DWB: variables, flujos e interfliujos*](https://www.gloh2o.org/mswep/)

## *Modelo Hydrologiska Byråns Vattenbalansavdelning - HBV*
El modelo HBV se desarrolló originalmente en el Instituto Meteorológico e Hidrológico Sueco (SMHI) para simular la relación precipitación-escorrentía y para la estimación hidrológica, y se utiliza en más de 30 países.La demanda computacional y los datos requeridos para el HBV son muy limitados, lo que le da preferencia para ser utilizado en una amplia gama de aplicaciones donde la obtención de datos es un problema.

El modelo consta de subrutinas para la interpolación meteorológica, la acumulación y el derretimiento de la nieve, la estimación de la evapotranspiración, un procedimiento de contabilización de la humedad del suelo, rutinas para la generación de escorrentía y, por último, un procedimiento de enrutamiento simple entre subcuencas y en lagos. Es posible ejecutar el modelo por separado para varias subcuencas y luego sumar las contribuciones de todas ellas.

<img src="https://drive.google.com/uc?export=view&id=1FMAy0rcEy3G4sO2Q1l7IgHGWdKbczNHd" alt="Drawing"  width="600" style="width: 100px;"/>

El metodo de humedad de suelo en WEAP es una conceptualización semiagregada de la hidrologia en cuencas unidimensional que representa la distribución de agua en la superficie de la cuenca a traves de dos estructuras, la zona radicular del suelo y la zona subterranea:

<img src="https://drive.google.com/uc?export=view&id=1Sq0F9y9EpFc5DaWcFnTCRiM2EFo1mK0b" alt="Drawing"  width="600" style="width: 100px;"/>



Este diagrama representa un comportamiento del agua en el suelo de la siguiente forma:
  $$ \frac{dx}{d} = \sum_{k=1}^{2} Q_k - \sum_{i=1}^{m} Q_i $$

Donde: representa el contenido de humedad en la superficie $h_o$

$$x \mapsto h_o(x, y, t)$$

En el diagrama previo, el suelo es representado por una zona radicular **reservorio 1** y una zona subterranea **reservorio 2**.

El primer reservorio trabaja bajo el siguiente esquema de balance entre ingresos y salidas:
$$
S_w \frac{dh_j}{dt} = P_e(t) - ET(t) - Q_s(t) - Q_{ss}(t) - Q_p(t)
$$

En esta ecuación, $S_w$ es la capacidad de almacenamiento en la zona radicular, $h_j$ representa la humedad del suelo en la zona radicular para la porcion de suelo/vegetacion $j$ de la cuenca, $P_e(t)$ la precipitación, $ET(t)$ la evapotranspiración del suelo y cobertura, $Q_s(t)$ la escorrentia directa,
$Q_{ss}(t)$ la escorrentia producida en la zona radicular y $Q_p(t)$ la percolacion hacia la zona subterranea.

Para el metodo de humedad de suelos de WEAP, esta expresion es definida de la siguiente forma, para el reservorio 1:

$$
S_w\frac{dh_{1j}}{dt} = Pe_{(t)} -ETo_{(t)}kc_{(t)}\left(\frac{5h_{1j} - 2(h_{1j})^2}{3}\right) - Pe_{(t)}h_{1j}^{RRF{j}}-f_j(ks_j)(h_{1j})^2 - (1-f_j)(ks_j)(h_{1j})^2
$$

Donde; $h_{1j}$ es la humedad del suelo correspondiente al reservorio 1 del modelo, $ETo_{(t)}$ es la evapotranspiracion de referencia, $kc_{(t)}$ es el coeficiente de cultivo/vegetación, $RRF{j}$ [adimensional] es el factor de resistencia al escurrimiento, con los valores inversamente proporcionales a las tasas más altas de escorrentía superficial (suelos sin cobertura vegetal). Este factor es proporcional al índice de área foliar LAI. $ks_j [L T^{-1}]$  es denominado  conductividad radicular. $f_j$ [adimensional] es un parámetro de calibración relacionado al tipo de suelo, cobertura y topografía, que define la proporción que escurre horizontalmente $(f_j)$ "interflujo" o verticalmente $(1 - f_j)$ "percolacion".

Para el reservorio 2, la ecuacion es la siguiente:

$$
Dw\frac{dh_{2j}}{dt} = \sum_{j=1}^{N} (1-f_j)ks_j h_{1_j} - ks_{2j} h_{2}^2
$$

Donde; $ks_{2j}$ es la conductividad en la zona subterranea (no tiene subindice $j$, por ser un factor general para toda la cuenca), $h_{2}$ es el contenido de agua en la zona subterranea, y $Dw$ es la capacidad de almacenamiento en la zona subterranea.

## *Temperatura*

```python
%%R -w 1200 -h 624 -u px

# Definir el periodo de analisis de los datos
fechas <- seq(from = as.Date("2000-01-01"), to = as.Date("2022-12-31"), by = "month")

# Asignar la propiedad de fechas a los archivos netcdf
TMed <- setZ(Tmed, fechas, name = "Time")
TMax <- setZ(Tmax, fechas, name = "Time")
TMin <- setZ(Tmin, fechas, name = "Time")

# Promedio mensual multianual
TMed.mensual <- zApply(TMed, by=format.Date(fechas, "%m"), fun= "mean", na.rm = F)
TMax.mensual <- zApply(TMax, by=format.Date(fechas, "%Y-%m"), fun= "mean", na.rm = F)
TMin.mensual <- zApply(TMin, by=format.Date(fechas, "%m"), fun= "mean", na.rm = F)

names(TMed.mensual) <- month.abb
plot(TMed.mensual, col = RColorBrewer::brewer.pal(8, "OrRd"))

# Promedio anual multianual
TMed.anual <- mean(zApply(TMed, by=format.Date(fechas, "%Y"), fun= "mean", na.rm = F), na.rm = TRUE)
plot(TMed.anual, col = RColorBrewer::brewer.pal(8, "OrRd"), title = "Temperatura [°C]")

# Crear una serie tiempo de temperatura maxima mensual
prom.Tmax <- cellStats(TMax.mensual, mean, na.rm = T)

plot(fechas, prom.Tmax, yaxt = "n", col = "darkgreen", ylab = "Temperatura [°C]", type = "l", title = "Temp. Máx. mensual")
axis(2, las =2)
```

## *Precipitación*

```python
%%R -w 1200 -h 624 -u px

# Definir las fechas de loas capas del netcdf
fechas.p <- seq(from = as.Date("1981-01-01"), to = as.Date("2024-03-31"), by = "month")
Prec <- setZ(Prec, fechas.p, name = "Time")

# Promedio mensual multianual
Prec.mensual <- zApply(Prec, by=format.Date(fechas.p, "%Y-%m"), fun= "sum", na.rm = F)
Prec.mensual.prom <- zApply(Prec.mensual, by=format.Date(fechas.p, "%m"), fun= "mean", na.rm = F)
names(Prec.mensual.prom) <- month.abb
Prec.mensual.agregado <- cellStats(Prec.mensual, mean, na.rm = TRUE)
Prec.anual <- zApply(Prec, by=format.Date(fechas.p, "%Y"), fun= "sum", na.rm = F)
Prec.anual.agregada <- cellStats(Prec.anual, mean, na.rm = TRUE)
Prec.anual.prom <- mean(Prec.anual, mean, na.rm = TRUE)

# Graficar los mapas de precipitación mensual y anual
plot(Prec.mensual.prom, col = RColorBrewer::brewer.pal(7, "Blues"))
plot(fechas.p, Prec.mensual.agregado, type = "l", col = "blue", yaxt = "n", ylab = expression(paste("Prec. [mm mes"^-1,"]",sep ="")))
axis(2, las =2)
plot(Prec.anual.prom, col= RColorBrewer::brewer.pal(7, "Blues"))
barplot(Prec.anual.agregada, col = RColorBrewer::brewer.pal(3, "Blues"), border = RColorBrewer::brewer.pal(3, "Blues"), las =2, ylab= expression(paste("Prec. [mm year"^-1,"]",sep ="")))
```
