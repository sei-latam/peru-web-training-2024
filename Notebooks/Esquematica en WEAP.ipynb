# Desarrollo del esquema de un modelo

WEAP permite desarrollar modelos que son trabajados con informacion de repositorios SIG en linea. Para realizar esto, utilizamos el modelo de la sesion previa. Para esto se debe ubicar en la opcion de "Schematic".

<img src="https://drive.google.com/uc?export=view&id=17kmR4wTQYfb2KGpnxn9aQA407Sx5rcN2" alt="Drawing"  width="800" style="width: 100px;"/>

Se abrira un menu en la derecha de WEAP con opcion de configuracion de datos online para la definicion de una esquematica en WEAP. En esta etapa, **por favor se pide borrar los elementos creados en la esquematica de la reunion previa**, vale decir, eliminar los dos nodos de demanda y el rio. WEAP puede **delinear automáticamente** cuencas hidrográficas y ríos (utilizando datos digitales de elevación), calcular la superficie (desagregada por banda de elevación y cobertura del suelo), descargar datos climáticos históricos para cada cuenca (por banda de elevación) y crear una capa de mapa de fondo de resumen climático.
A continuacion se muestra un esquema con la descripción de las caracteristicas de este menu, para conectar bases de datos externas para el desarrollo de un modelo en WEAP.
<img src="https://drive.google.com/uc?export=view&id=1FQ3sddAemT6MPjiWCOVKiMlncugfbOWT" alt="Drawing"  width="600" style="width: 100px;"/>

Para crear cuencas con la asistencia de WEAP, se puede hacer uso de la esquematica. Con el mouse, la dinamica de creacion de cuencas consiste en hacer click izquierdo en el area de aporte de la cuenca que se desea delimitar. una vez ejecuta aparece una ventana emergente en la cual se debe definir el nombre de la cuenca.
